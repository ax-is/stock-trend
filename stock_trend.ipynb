```python
# Import necessary libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Function to fetch and analyze stock data
def stock_trend_analysis(ticker, start_date, end_date):
    # Fetch stock data using yfinance
    stock_data = yf.download(ticker, start=start_date, end=end_date)

    if stock_data.empty:
        print("No data found for the given ticker and date range.")
        return

    # Display basic statistics
    print(f"Stock: {ticker}")
    print("Basic Statistics:")
    print(stock_data['Close'].describe())

    # Plot stock closing price trend
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data['Close'], label=f'{ticker} Closing Price')
    plt.title(f"{ticker} Stock Price Trend ({start_date} to {end_date})")
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.legend()
    plt.grid()
    plt.show()

# User input for ticker and date range
print("Simple Stock Trend Analysis Tool")
ticker = input("Enter the stock ticker (e.g., AAPL for Apple): ")
start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

# Call the function
stock_trend_analysis(ticker, start_date, end_date)
```
